```python
## Creating Traing and Testing Dataset Using ImageGenerator 
from tensorflow.keras.preprocessing.image import ImageDataGenerator 
```

```python
trainDataGen=ImageDataGenerator(rescale=1./255,
                                shear_range=0.2,
                                zoom_range=0.2,
                                horizontal_flip=True
                            )
```

```python
testDataGen=ImageDataGenerator(rescale=1./255)
```

```python
training_set=trainDataGen.flow_from_directory('Datasets/Train',
                                              target_size=(227,227),
                                              batch_size=400,
                                              class_mode='categorical'
                                                )
```

```python
test_set=testDataGen.flow_from_directory('Datasets/test',
                                              target_size=(227,227),
                                              batch_size=400,
                                              class_mode='categorical'
                                                )
```

```python
## Start Creating The Layers 
```

```python
AlexNet=Sequential()
AlexNet.add(Conv2D(filters=96,kernel_size=11,activation='relu',input_shape=(227,227,3),strides=4,padding='valid'))
AlexNet.add(MaxPooling2D(pool_size=3,strides=2,padding='valid'))
AlexNet.add(BatchNormalization())
AlexNet.add(Conv2D(filters=256,kernel_size=5,activation='relu',strides=1,padding='same'))
AlexNet.add(MaxPooling2D(pool_size=3,strides=2,padding='valid'))
AlexNet.add(BatchNormalization())
AlexNet.add(Conv2D(filters=384,kernel_size=3,activation='relu',strides=1,padding='same'))
AlexNet.add(Conv2D(filters=384,kernel_size=3,activation='relu',strides=1,padding='same'))
AlexNet.add(Conv2D(filters=256,kernel_size=3,activation='relu',strides=1,padding='same'))
AlexNet.add(MaxPooling2D(pool_size=2,strides=2,padding='valid'))
AlexNet.add(Dropout(0.4))
AlexNet.add(Flatten())
AlexNet.add(Dense(units=4096,activation='relu'))
AlexNet.add(Dense(units=1000,activation='relu'))
AlexNet.add(Dense(units=2,activation='softmax'))
```

```python
AlexNet.summary()
```

```python
AlexNet.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
AlexNet.fit(x=training_set,validation_data=test_set,epochs=5)
```
